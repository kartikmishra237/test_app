```python
df_enc=pd.get_dummies(df, columns=['Gender','Customer Type','Type of Travel','Class','Satisfaction'],dtype=int,drop_first=True)
df_enc
```

```python
plt.figure(figsize=(25, 25))
co_mtx = df_enc.corr(numeric_only=True)
sns.heatmap(co_mtx, cmap="YlGnBu", annot=True)
```

```python
df_enc['Flight Distance'].describe()
```

```python
#Corrected using error statements 
df_enc['Flight Distance']=df_enc['Flight Distance'].apply(lambda x:'3655' if x=='3655$' else x)
df_enc['Flight Distance']=df_enc['Flight Distance'].apply(lambda x:'2257' if x=='22%7' else x)
df_enc['Flight Distance']=df_enc['Flight Distance'].apply(lambda x:'2527' if x=='2$27' else x)
```

```python
df_enc['Flight Distance'] = df_enc['Flight Distance'].astype(int)
```

```python
#Filling Null values in Arrival Delay
mean=df_enc['Arrival Delay'].mean()
df_enc['Arrival Delay'] = df_enc['Arrival Delay'].fillna(mean)
df_enc
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
x=df_enc.drop(['Satisfaction_Satisfied'],axis=1)
y=pd.DataFrame(df_enc['Satisfaction_Satisfied'])
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
df_scaled = scaler.fit_transform(x)
x_sc=pd.DataFrame(df_scaled)
x_sc
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_sc,y,test_size=0.2,random_state=42,stratify=y)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier , VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
#LogisticRegression
logit=LogisticRegression()
logit=logit.fit(x_train,y_train)
y_pred_lr=logit.predict(x_test)

accuracy_lr=accuracy_score(y_test,y_pred_lr)
precision_lr=precision_score(y_test, y_pred_lr)
recall_lr=recall_score(y_test, y_pred_lr)
f1_lr=f1_score(y_test, y_pred_lr)
cm_lr=confusion_matrix(y_test, y_pred_lr)

print('Accuracy:',accuracy_lr)
print()
print('Precision:',precision_lr)
print()
print('Recall:',recall_lr)
print()
print('F1_Score:',f1_lr)
print()
print('Confusion_Matrix:\n',cm_lr)
```

```python
#DTree
dtree=DecisionTreeClassifier()
dtree=dtree.fit(x_train,y_train)
y_pred_dt=dtree.predict(x_test)
accuracy_dt=accuracy_score(y_test,y_pred_dt)
accuracy_dt

accuracy_dt=accuracy_score(y_test,y_pred_dt)
precision_dt=precision_score(y_test, y_pred_dt)
recall_dt=recall_score(y_test, y_pred_dt)
f1_dt=f1_score(y_test, y_pred_dt)
cm_dt=confusion_matrix(y_test, y_pred_dt)

print('Accuracy:',accuracy_dt)
print()
print('Precision:',precision_dt)
print()
print('Recall:',recall_dt)
print()
print('F1_Score:',f1_dt)
print()
print('Confusion_Matrix:\n',cm_dt)
```

```python
#RandomForest
rf=RandomForestClassifier()
rf=rf.fit(x_train,y_train)
y_pred_rf=rf.predict(x_test)
accuracy_rf=accuracy_score(y_test,y_pred_rf)
accuracy_rf

accuracy_rf=accuracy_score(y_test,y_pred_rf)
precision_rf=precision_score(y_test, y_pred_rf)
recall_rf=recall_score(y_test, y_pred_rf)
f1_rf=f1_score(y_test, y_pred_rf)
cm_rf=confusion_matrix(y_test, y_pred_rf)

print('Accuracy:',accuracy_rf)
print()
print('Precision:',precision_rf)
print()
print('Recall:',recall_rf)
print()
print('F1_Score:',f1_rf)
print()
print('Confusion_Matrix:\n',cm_rf)
```

```python
#KNN
knn=KNeighborsClassifier(n_neighbors=5)
knn=knn.fit(x_train,y_train)
y_pred_knn=knn.predict(x_test)
accuracy_knn=accuracy_score(y_test,y_pred_knn)
accuracy_knn

accuracy_knn=accuracy_score(y_test,y_pred_knn)
precision_knn=precision_score(y_test, y_pred_knn)
recall_knn=recall_score(y_test, y_pred_knn)
f1_knn=f1_score(y_test, y_pred_knn)
cm_knn=confusion_matrix(y_test, y_pred_knn)

print('Accuracy:',accuracy_knn)
print()
print('Precision:',precision_knn)
print()
print('Recall:',recall_knn)
print()
print('F1_Score:',f1_knn)
print()
print('Confusion_Matrix:\n',cm_knn)
```

```python
#GNB
gnb=GaussianNB()
gnb=gnb.fit(x_train,y_train)
y_pred_gnb=gnb.predict(x_test)
accuracy_gnb=accuracy_score(y_test,y_pred_gnb)
accuracy_gnb

accuracy_gnb=accuracy_score(y_test,y_pred_gnb)
precision_gnb=precision_score(y_test, y_pred_gnb)
recall_gnb=recall_score(y_test, y_pred_gnb)
f1_gnb=f1_score(y_test, y_pred_gnb)
cm_gnb=confusion_matrix(y_test, y_pred_gnb)

print('Accuracy:',accuracy_gnb)
print()
print('Precision:',precision_gnb)
print()
print('Recall:',recall_gnb)
print()
print('F1_Score:',f1_gnb)
print()
print('Confusion_Matrix:\n',cm_gnb)
```
